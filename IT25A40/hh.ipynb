```python
StMarks.sort_values(["Obtained marks"],ascending=0)
```

```python
StMarks["Percentage"]=(StMarks["Obtained marks"]*100/StMarks["Total Marks"])
print(StMarks)
```

```python
# a-80-100
# b-60-80
# c-40-60
# d-40-20

def assign_grade(Percentage):
    if Percentage >= 80 and Percentage <=100:
        return 'A'
    elif Percentage >= 60 and Percentage <=80:
        return 'B'
    elif Percentage >= 40 and Percentage <=60:
        return 'C'
    elif Percentage >= 40 and Percentage <=20:
        return 'D'
    else :
        return 'F'
StMarks["Grade"]=StMarks['Percentage'].apply(assign_grade)
StMarks
```
