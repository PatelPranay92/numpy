```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = [0,2,4,6,8]
y = [0,4,16,36,40]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'o') #markers
```

```python
plt.plot(x,y,'*') #markers
```

```python
plt.plot(x,y,'x') #markers
```

```python
plt.plot(x,y,'p') #markers
```

```python
plt.plot(y) #simple line 
```

```python
plt.plot(x,y,'x',linestyle='-') #line Style
```

```python
plt.plot(x,y,'x',linestyle='-.') #line Style
```

```python
plt.plot(x,y,'x',linestyle=':') #line Style
```

```python
plt.plot(x,y,'x',linestyle='dotted') #line Style
```

```python
plt.plot(x,y,'x',linestyle='dotted',color='r') #Color red
```

```python
plt.plot(x,y,'x',linestyle='dotted',color='k') #color black
```

```python
plt.plot(x,y,'*',linestyle='dotted',color='k',ms=10) #Marker size
```

```python
plt.plot(x,y,'*',linestyle='dotted',color='k',ms=10,mec='r') # marker edge color
```

```python
plt.plot(x,y,'*',linestyle='dotted',color='k',ms=10,mec='r',mfc='y') # marker inside color
```

```python
plt.plot(x,y,'*',linestyle='dotted',color='k',ms=10,mec='r',mfc='y',lw='10') #line width
```

```python
plt.plot(x,y,'*',linestyle='dotted',color='k',ms=10,mec='r',mfc='y',lw='10')
plt.title("line chart Demo")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
plt.plot(x,y,'*',linestyle='dotted',color='k',ms=10,mec='r',mfc='y',lw='10')
font1={'family':'serif','color':'b','size':20} #dic for font 
font2={'family':'serif','color':'r','size':10}
plt.title("line chart Demo",fontdict=font1,loc='left') #location of title
plt.xlabel("X-axis",fontdict=font2)
plt.ylabel("Y-axis",fontdict=font2)
```
