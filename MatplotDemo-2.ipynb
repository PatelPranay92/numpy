```python
import pandas as pd

df = pd.read_csv('students.csv')
df['Total Marks'] = df['Math Marks'] + df['Science Marks'] + df['English Marks']
print(df)
```

```python
student = df["Student Name"]
marks = df["Total Marks"]
plt.pie(marks,labels=student)
plt.show()
```

```python
student = df["Student Name"]
marks = df["Total Marks"]
plt.bar(student,marks)
plt.show()
```

```python
student = df["Student Name"]
marks = df["Total Marks"]

def show_marks(pct):
    total = sum(marks) 
    value = int(pct * total / 100)  
    return f'{value}'

plt.pie(marks, labels=student, autopct=show_marks)

plt.show()

```
